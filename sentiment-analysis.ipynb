```python
import transformers
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier("Hey I'm here to learn transformers, how to get in to it?")
```

```python
results = classifier(["I just wanna learn something beyond to deep learning","Otherwise I will stop learning machine learning I took a transformers course"])
for result in results:
  print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```
